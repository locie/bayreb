# Parameter estimation in a deterministic 2R2C model

This tutorial is the implementation of the example given in the review paper published in *Energy and Buildings*:

[Rouchier S. (2018) Solving inverse problems in building physics: an overview of guidelines for a careful and optimal use of data. *Energy and Buildings*, DOI:10.1016/j.enbuild.2018.02.009](https://www.sciencedirect.com/science/article/pii/S0378778817317942)

## Introduction

A test house has been monitored experimental measurements of the ambient (outdoor) temperature $T_a$, indoor temperature $T_i$ and internal heat input $q$ are available, and the target of the exercise is to identify the value of thermal resistors and capacitors that describe the building energy performance. These parameters form an unknown vector $\theta$. A 2R2C model is selected, that simulates the indoor temperature $T_i$, along with an unobserved envelope temperature $T_e$.

```python
from IPython.display import Image
Image(filename='2R2C_model.png', width = 300)
```

The 2R2C model shown here can be summarised by the following system of equations:

\begin{equation}
	\begin{bmatrix} \dot{T}_e(t) \\ \dot{T}_i(t) \end{bmatrix} = \begin{bmatrix} -\dfrac{1}{R_1C_1}-\dfrac{1}{R_2C_1} & \dfrac{1}{R_2C_1} \\ \dfrac{1}{R_2C_2} & -\dfrac{1}{R_2C_2} \end{bmatrix} \begin{bmatrix} T_e(t) \\ T_i(t) \end{bmatrix} + \begin{bmatrix} \dfrac{1}{R_1C_1} & 0 \\ 0 & \dfrac{1}{C_2} \end{bmatrix} \begin{bmatrix} T_a(t) \\ q(t) \end{bmatrix} 
\end{equation}

The target of the exercise is to use the data to estimate $\theta$.

## The data

First, let us load some packages and functions that we will use later.

```python
import pandas
import numpy as np
from scipy.linalg import expm
from numpy.linalg import inv

def dot3(A,B,C):
    """ Useful function that computes the dot product of three matrices"""
    return np.dot(A, np.dot(B,C))
```

The data are contained within the `data01.csv` file, available on the BAYREB github repository in the same folder as this notebook. Upon reading this file, we can store each of its column in a separate numpy array for convenience. We can then take a look at it. Note that this section requires the [seaborn library](https://seaborn.pydata.org/), which facilitates making nice looking graphs.

```python
dataset = pandas.read_csv('data01.csv')
time_ = np.array(dataset['Time'])
T_in  = np.array(dataset['T_int'])  # indoor temperature
T_ext = np.array(dataset['T_ext'])  # outdoor temperature
q     = np.array(dataset['q'])      # indoor prescribed heat

%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid", rc={'font.family':'serif'})
sns.set_context('paper', font_scale=1.5)
palette = sns.color_palette()

fig = plt.figure(figsize=(7, 4))
ax1 = fig.add_subplot(111)
ax1.plot(time_/34/3600, T_in , color = palette[2], label = '$T_i$')
ax1.plot(time_/34/3600, T_ext, color = palette[0], label = '$T_a$')
ax2 = ax1.twinx()
ax2.grid(None)
ax2.plot(time_/34/3600, q, color = palette[3], label = '$q$')

ax1.set_xlabel('Time (days)')
ax1.set_ylabel('Indoor and outdoor temperature')
ax2.set_ylabel('Heating power $q$', color = palette[3])
ax1.legend(loc = 'upper left')
ax2.legend(loc = 'upper right')
```

The monitoring period is roughly 2.6 days, during which a 2 kW heat input is imposed in the building during 24h. The indoor temperature increases nicely during this heating period, then naturally decreases after the heater is turned off. This is the observation sequence used to train our model.

## The model

In this exercise, we only consider a deterministic model: system error is overlooked, and the only difference between model predictions and observations is measurement noise. The equations that describe the temporal evolution of indoor and envelope temperatures and written above, given the value of thermal resistors and capacitors.

The following block of code writes this model in a discrete form, that returns $T_i$ as a function of these parameters, with the same time discretisation as is given by the data. More details on the content of this function are available in [the paper](https://www.sciencedirect.com/science/article/pii/S0378778817317942).

```python
delta_t = time_[1] - time_[0]       # time step size
u = np.vstack([T_ext ,q]).T     # array of all inputs

def RC_model_simulation(time_, R1, R2, C1, C2, xe_0):
    
    # Matrices of the system in continuous form
    Ac = np.array([[-1/(C1*R1)-1/(C1*R2), 1/(C1*R2)],
                   [1/(C2*R2), -1/(C2*R2)]])
    Bc = np.array([[1/(C1*R1), 0],
                   [0, 1/C2]])
    
    # Matrices of the discretized state-space model
    F = expm(Ac*delta_t)
    G = dot3(inv(Ac), F-np.eye(2), Bc)
    H = np.array([[0, 1]])
    
    # Initialisation of the states
    x = np.zeros((len(time_), 2))
    x[0] = np.array((xe_0, T_in[0]))
    
    # Simulation
    for i in range(1,len(time_)):
        x[i] = np.dot(F, x[i-1]) + np.dot(G, u[i-1])
    
    # This function returns the second simulated state only
    return np.dot(H, x.T).flatten()
```

In the above code, `xe_0` stands for the initial value of the "envelope temperature" $T_e$. It is considered unknown and will be inferred along with the other parameters.

Now that the evaluation function is defined, we can call one of the built-in curve fitting methods from scipy. We do so by providing an initial guess for each of the unknown parameters: this list should be defined in the same order as the arguments of the evaluation function. Resistances $R_1$ and $R_2$ have a starting value of 10$^{-2}$ K/W, capacitances $C_1$ and $C_2$ start at 10$^7$ J/K and $T_{e,0}$ at 20°C.

```python
from scipy.optimize import curve_fit

theta_init = [1e-2, 1e-2, 1e7, 1e7, 20]
popt, pcov = curve_fit(RC_model_simulation,
                       xdata = time_,
                       ydata = T_in,
                       p0 = theta_init,
                       method='lm')

# Standard deviation of the parameter estimates
stdev = np.diag(pcov)**0.5
# Correlation matrix
R = dot3( np.linalg.inv(np.diag(stdev)), pcov, np.linalg.inv(np.diag(stdev)))
# t-statistic
t_stat = popt / stdev
```

The `curve_fit` method from scipy returns:

* the optimal value of parameters `popt`, which is the least-square estimator $\hat{\theta}_{LS}$
* the covariance matrix `pcov` from which the standard deviation of parameter estimates $\sigma_{\hat{\theta}}$ can be calculated

Results are summarised in the following table. You should obtain similar values if you run the script.

We can also make check how well our calibrated model fits the data, by plotting the observed indoor temperature along with the temperature predicted using optimised parameters:

```python
# Calculating the indoor temperature predicted with the optimal parameters
T_in_opt = RC_model_simulation(time_, popt[0], popt[1], popt[2], popt[3], popt[4])

plt.figure(figsize=(5, 3))
plt.plot(time_/24/3600, T_in, '-k', linewidth=1.5, label='Observation')
plt.plot(time_/24/3600, T_in_opt, '-r', linewidth=1.5, label='Simulation')
plt.xlabel('Time (days)')
plt.ylabel('Indoor temperature')
plt.legend()
```
